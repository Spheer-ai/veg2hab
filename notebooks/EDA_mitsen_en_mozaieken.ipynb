## Beperkende criteria

```python
dt['beperkende criteria'].nunique()
```

```python
bc = dt['beperkende criteria'].str.split('mits').explode().reset_index(drop=True)
```

```python
bc = bc.str.strip()
bc = bc.str.replace(' en$', '', regex=True)
bc = bc[bc != '']
bc = bc[~bc.isna()]
```

```python
bc.shape
```

```python
bc.nunique()
```

```python
bc.to_excel('../data/beperkende_criteria.xlsx', index=False, header=True)
```
